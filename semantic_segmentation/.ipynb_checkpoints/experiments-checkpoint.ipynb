```python
import utils
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = utils.load_image("./media/test_mask.png")
iid = utils.load_image("./media/test_iid.png")
testlid = utils.load_image("./media/test_lid.png")
```

```python
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
color = np.array([128, 64,128])
 
# Here we are defining range of bluecolor in HSV
# This creates a mask of blue coloured 
# objects found in the frame.
mask = cv2.inRange(hsv, color, color)
 
# The bitwise and of the frame and mask is done so 
# that only the blue coloured objects are highlighted 
# and stored in res
res = cv2.bitwise_and(image,image, mask= mask)
plt.imshow(np.hstack([hsv, res]))
plt.show()
```
